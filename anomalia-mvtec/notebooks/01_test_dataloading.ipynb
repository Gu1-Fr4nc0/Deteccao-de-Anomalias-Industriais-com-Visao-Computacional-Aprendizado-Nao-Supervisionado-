```python
import matplotlib.pyplot as plt
from torchvision import transforms
from torch.utils.data import DataLoader
from src.dataset import MVTecDataset

# Configurações
DATA_PATH = './data/mvtec_anomaly_detection' # Ajuste se sua pasta for diferente
CATEGORY = 'screw'

# Transformações básicas (Autoencoder precisa de tamanho fixo)
data_transforms = transforms.Compose([
    transforms.Resize((128, 128)), # Reduz para ficar leve
    transforms.ToTensor(),
])

# Instancia os Datasets
train_ds = MVTecDataset(DATA_PATH, CATEGORY, data_transforms, mode='train')
test_ds = MVTecDataset(DATA_PATH, CATEGORY, data_transforms, mode='test')

print(f"Imagens de Treino (Só Good): {len(train_ds)}")
print(f"Imagens de Teste (Good + Bad): {len(test_ds)}")

# Visualiza um exemplo
img, label = train_ds[0]
print(f"Shape da imagem: {img.shape}") # Deve ser [3, 128, 128]
print(f"Label (deve ser 0): {label}")

plt.imshow(img.permute(1, 2, 0)) # Converte de [C, H, W] para [H, W, C] pro plot
plt.title("Exemplo do Dataset - Screw")
plt.show()
```
